```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", error_bad_lines=False)

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years
```

```python
df.head(10)
```

With your dataframe set up, time to start creating visualizations!

```python
def count(series):
    if series>0:
        return 1

df['count'] = df['bookID'].apply(count)
df['count'].describe()
```

```python
# Use your Python skills to create a visualization here.
#Number of books published per year.
df_byyear = df.groupby('year').sum().sort_values('year', ascending=False)
df_byyear = df_byyear.reset_index()
df_byyear.head(21)

df_byyear_recent = df_byyear[df_byyear['year'] >= '2000']
df_byyear_recent.head(21)

# sb.lineplot(data=df_byyear_recent, x='year', y='count')

plt.plot('year', 'count', data=df_byyear_recent)
plt.title('Books Published by Year', color='black')
plt.xlabel('Year', color='black')
plt.ylabel('Books Published', color='black')
plt.xticks(color='black', rotation = 45)
plt.yticks(color='black')
plt.savefig('linechart_single.png')
```
